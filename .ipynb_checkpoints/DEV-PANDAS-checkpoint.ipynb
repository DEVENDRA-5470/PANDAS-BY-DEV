```python
#8. Merging and Reshaping##
# `pandas.concat()`: Concatenate DataFrames.
# `pandas.merge()`: Merge DataFrames based on a common key.
# `df.pivot()`: Pivot a DataFrame.
# `df.melt()`: Unpivot a DataFrame.


######## `pandas.concat()`: Concatenate DataFrames.

import pandas as pd

# Create two DataFrames
df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Age': [24, 27]
})

df2 = pd.DataFrame({
    'Name': ['Charlie', 'David'],
    'Age': [22, 25]
})

#===> Example 1: Concatenating DataFrames Vertically (axis=0)
result = pd.concat([df1, df2], ignore_index=True)
result

#===> Example 2: Concatenating DataFrames Horizontally (axis=1)
result = pd.concat([df1, df2], axis=1)
result

#===> Example 3: Concatenate with Different Columns (Using join='outer')
result = pd.concat([df1, df2], axis=0, join='outer')
result

#===> Example 4: Resetting the Index (Using ignore_index=True)
result = pd.concat([df1, df2], axis=0, ignore_index=True)
result

#===> Example 5: Concatenate with Keys (Creating a Hierarchical Index)
result = pd.concat([df1, df2], axis=0, keys=['Group1', 'Group2'])
result
```

```python
########## `pandas.merge()`: Merge DataFrames based on a common key.
import pandas as pd

# Create two sample DataFrames
df1 = pd.DataFrame({
    'PROD_ID': [1, 2, 3, 4],
    'Product': ['Apple', 'Banana', 'Orange', 'Grapes']
})

# Sample data for df2
df2 = pd.DataFrame({
    'CUS_PROD_ID': [2, 3, 5],
    'Customer': ['John', 'Alice', 'Bob']
})

# Print column names to ensure they are correct
print("df1 columns:", df1.columns)
print("df2 columns:", df2.columns)

# Example 1: Inner Join (Default) - using correct column names
result = pd.merge(df1, df2, how='inner', left_on='PROD_ID', right_on='CUS_PROD_ID')
print("\nInner Join:")
print(result)

# Example 2: Left Join
result = pd.merge(df1, df2, how='left', left_on='PROD_ID', right_on='CUS_PROD_ID')
print("\nLeft Join:")
print(result)

# Example 3: Right Join
result = pd.merge(df1, df2, how='right', left_on='PROD_ID', right_on='CUS_PROD_ID')
print("\nRight Join:")
print(result)

# Example 4: Outer Join
result = pd.merge(df1, df2, how='outer', left_on='PROD_ID', right_on='CUS_PROD_ID')
print("\nOuter Join:")
result
```

```python
####### `df.pivot()`: Pivot a DataFrame.
import pandas as pd

# Create a sample DataFrame
data = {
    'City': ['New York', 'New York', 'Los Angeles', 'Los Angeles', 'Chicago', 'Chicago'],
    'Month': ['Jan', 'Feb', 'Jan', 'Feb', 'Jan', 'Feb'],
    'Temperature': [30, 32, 65, 68, 22, 25],
    'Humidity': [8.0, 7.8, 5.0, 5.2, 6.0, 6.2]
}

df = pd.DataFrame(data)

#===> Example 1: Pivot with Multiple Columns and Single Value Column
# Pivot the DataFrame with City as the index and Month as columns
pivot_df = df.pivot(index='City' , columns='Month', values='Temperature')
pivot_df

#===> Example 2: Pivot with Duplicate Entries
pivot_df = df.pivot(index='City', columns='Month', values=['Temperature', 'Humidity'])
pivot_df

#=====

data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    'Product': ['Apple', 'Banana', 'Apple', 'Banana'],
    'Region': ['North', 'South', 'North', 'South'],
    'Sales': [100, 150, 200, 250]
}

df = pd.DataFrame(data)


#===> Example 3: Pivot with Index and Multiple Columns.
pivot_df = df.pivot(index='Date', columns=['Product', 'Region'], values='Sales')
pivot_df

```

```python
##### `df.melt()`: Unpivot a DataFrame.


import pandas as pd

# Create a sample DataFrame
data = {
    'City': ['New York', 'Los Angeles', 'Chicago'],
    'Population': [8419600, 3980400, 2716000],
    'Area': [783.8, 503.0, 606.1],
    'GDP': [1700, 1000, 600]
}

df = pd.DataFrame(data)

#===> Example 1: Basic Usage
# Melt only the 'Population' and 'Area' columns
melted_df = df.melt(id_vars=['City'], value_vars=['Population', 'Area'], var_name='Metric', value_name='Value')

melted_df


# Sample DataFrame with multiple identifier columns
data = {
    'Region': ['North', 'North', 'South', 'South'],
    'Product': ['A', 'B', 'A', 'B'],
    'Sales_Q1': [100, 200, 150, 250],
    'Sales_Q2': [110, 210, 160, 260]
}
df = pd.DataFrame(data)

# Melt the DataFrame with multiple identifier variables
melted_df = df.melt(id_vars=['Region', 'Product'], value_vars=['Sales_Q1', 'Sales_Q2'], var_name='Quarter', value_name='Sales')

melted_df
```

```python
#9. Time Series##
# `pandas.to_datetime()` converts data into datetime format.
# `df.resample()` resamples time series data based on specified frequency.
# `df.shift()` shifts values by a specified number of periods.


# `pandas.to_datetime()` converts data into datetime format.
import pandas as pd
# Sample date string
date_str = '2024-12-25 '
# Convert to datetime
date = pd.to_datetime(date_str)
print(date)

import pandas as pd
# Sample list of date strings
date_list = ['2024-12-25', '2024-12-26', '2024-12-27']

# Convert to datetime
dates = pd.to_datetime(date_list)
print(dates)


import pandas as pd
# Sample DataFrame with date strings
data = {'Date': ['2024-12-25', '2024-12-26', '2024-12-27']}
df = pd.DataFrame(data)
# Convert the 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])
df



### `df.resample()` resamples time series data based on specified frequency.
import pandas as pd
# Sample data with dates
data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-02-01', '2024-02-02'],
    'Sales': [100, 150, 200, 250, 300]
}

# Create a DataFrame
df = pd.DataFrame(data)
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
# Set 'Date' column as the index
df.set_index('Date', inplace=True)
# Resample data to monthly frequency and calculate the mean sales for each month
monthly_sales = df.resample('ME').sum()
monthly_sales = df.resample('2D').sum()
# monthly_sales = df.resample('Y').mean()
monthly_sales


### `df.shift()` shifts values by a specified number of periods.
import pandas as pd
# Sample data with a datetime index
dates = pd.date_range('2024-01-01', periods=5, freq='D')
data = {'Sales': [100, 150, 200, 250, 300]}
# Create a DataFrame with datetime index
df = pd.DataFrame(data, index=dates)
# Shift the values by 1 day
df_shifted_time = df.shift(periods=1, freq='D')
print(df_shifted_time)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# `df.plot()`: Generate simple plots using Matplotlib.
# Read the CSV file
path = r"C:\Users\dev\OneDrive\Documents\EWR.csv"
frame = pd.read_csv(path)

# Set up colors: green for fees paid == 25000, red for others
colors=[]
for grade_level in frame['grade_level']:
    if grade_level == 8:
        colors.append('green')
    elif grade_level <= 4:
        colors.append('red')
    else:
        colors.append('blue')


# Plot a bar graph showing fees paid by each student with custom colors
frame.plot(x='first_name', y='grade_level', kind='bar', title='Fees Paid by Students', figsize=(12, 8), color=colors)

# Customize the plot
plt.xlabel('Student First Name')
plt.ylabel('Fees Paid')
plt.xticks(rotation=60, ha="right")  # Rotate the x-axis labels for better readability

# Show the plot
plt.show()
```

```python
# 2. Bar Plot
# Use Case: Compare the performance of different categories in a survey or business.
categories = ['January', 'February', 'March', 'April', 'May', 'June']
values = [23, 45, 56, 78, 80, 65]

c=['green' if v > 65 else 'yellow' if v > 45 else 'red'  for v in values]
plt.bar(categories, values, color=c)
plt.title("Monthly Sales Data")
plt.xlabel("Month")
plt.ylabel("Sales (in thousands)")
plt.show()
```

```python
# 3. Scatter Plot
# Use Case: Examine the relationship between advertising spend and product sales.
# Data: Number of hours studied vs Test Scores
import matplotlib.pyplot as plt
hours_studied = [1, 2, 3, 4, 5]
test_scores = [50, 55,60, 61, 67]

# Create a scatter plot
c=['green' if t > 55 else 'red' for t in test_scores]
plt.scatter(hours_studied, test_scores, color=c)

# Add title and labels
plt.title("Scatter Plot: Hours Studied vs Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Test Scores")
plt.grid(True)
# Display the plot
plt.show()
```

```python
# 4. Histogram
# Use Case: Explore the distribution of exam scores for a class.
import numpy as np
import matplotlib.pyplot as plt

# Generate exam scores using a normal distribution
mean_score = 70  # Mean score (average score)
std_deviation = 10  # Standard deviation
num_students = 1000  # Number of students

# Generate random exam scores following a normal distribution
exam_scores = np.random.normal(mean_score, std_deviation, num_students)

# Plot the histogram
plt.hist(exam_scores, bins=30, color='purple', edgecolor='black')

# Add title and labels
plt.title("Histogram of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Number of Students")

# Show the plot
plt.show()
```
